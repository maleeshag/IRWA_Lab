#Q3)Python implementation to print first h1 tag from https://www.sliit.lk

Beautiful Soup is a popular Python library used for web scraping and parsing HTML and XML documents. It provides a convenient way to extract data from web pages by navigating and searching the HTML structure.

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

#html content will be stored in this 'html' variable
html=urlopen("https://www.sliit.lk/")

#create an object and pass the html variable into that , to parse

bs=BeautifulSoup(html)

titles=bs.find_all(['h1'])

print("List of all the header tags: ",titles,sep='\n\n')
```

```python
#Second Method
import requests
f = requests.get("https://www.sliit.lk/")

soup = BeautifulSoup(f.content)

header = soup.find_all(["h1"])

print(header)
```

#Q4)How to print all the header tags from https://www.sliit.lk.

```python
html = urlopen("https://www.sliit.lk/")

#creating an object from the library BeautifulSoup
bs = BeautifulSoup(html)   # parsing the html content

titles = bs.find_all(["h1","h2","h3","h4","h5","h6"])

for h in titles:
  print(h)

  #for better readability
  print()
```

#Q5)Python implementation to retrieve all the links in a particular URL

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

#sending an http request Get, to the specified URL

html = urlopen("https://www.sliit.lk/")

#Create an object from beautifulsoup

bsObj = BeautifulSoup(html)

#iterating through all anchor(<a>) elements within the Parsed HTML
for link in bsObj.findAll("a"):
  #Checking whether we have 'href' attribute in the current anchor elements
  if 'href' in link.attrs:
    print(link.attrs["href"])
```
